> There are three synpotic patterns of circulation in the Santa Barbara Channel: upwelling, convergent, and relaxation. The basic ideas of this were laid out in [Harms & Winant, 1998](https://agupubs.onlinelibrary.wiley.com/doi/abs/10.1029/97JC02393) (see Fig. 10) and then considered in further detail by [Winant et al 2013](https://agupubs.onlinelibrary.wiley.com/doi/full/10.1029/2001JC001302) (see Fig. 5, top row). (The cyclonic eddy in the 1998 paper is termed convergent in the 2003 paper). 
> 
> Summarize these three states in a short narrative, including the typical season that these flow states tend to most frequently occur.

*Describe the three circulation patterns here.*

**Upwelling:**
>
Usually consists of equatorward flow from surface level to 45m depth, except in the area in Santa Barbara County that is adjacent to the main coastline. This pattern occurs mostly during Spring,  March and April, when winds are upwelling avorable and pressure at the surface is constant. 
>
**Convergent:**
>
Occurs during Summer, July-September, and consists of equatorward surface flow, and poleward flow at 45m depth. As a result of the flow convergence, the water body moves mostly in a offshore or westward direction. Winds during this pattern are strong and upwelling favorable and the variable upwelling has created surface pressure gradients alongshore. 
>
**Relaxation:**
> Often occuring during Winter, November-December
